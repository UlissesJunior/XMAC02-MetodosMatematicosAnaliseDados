1) Exiba histogramas lado a lado mostrando o índice de corrupção na Europa e na África. Use bins=15.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('CPI_2014.csv', encoding='latin-1')
df1 = df[df['Continent'] == 'Europe']
df2 = df[df['Continent'] == 'Africa']
```

```python
plt.rcParams['figure.figsize'] = [10, 6]
sns.set_style('darkgrid')

plt.subplot(1,2,1)
sns.histplot(df1['CPI_2014'], bins = 15, kde=False)
plt.title('Indíce de percepção de corrupção na Europa')
plt.xlabel('Nível de corrupção')
plt.ylabel('Quantidade')
plt.axvline(60, color='k', linestyle='--')

plt.subplot(1,2,2)
sns.histplot(df2['CPI_2014'], bins = 15, kde=False)
plt.title('Indíce de percepção de corrupção na África')
plt.xlabel('Nível de corrupção')
plt.ylabel('Quantidade')
plt.axvline(60, color='k', linestyle='--')

plt.show()
```
