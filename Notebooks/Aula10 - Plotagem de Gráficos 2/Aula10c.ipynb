## Gráfico de Barras
Vamos construir um gráfico de barras usando os dados do dataset Ebola.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('ebola.csv', encoding='latin-1')
```

```python
df
```

```python
plt.rcParams['figure.figsize'] = [8,6]
sns.barplot(x='Type', y='Death', hue='Countries', data=df, errorbar=None)
plt.title('Mortalidade do Ebola nos países Serra Leoa e Guiné')
plt.xlabel('Tipos de casos')
plt.ylabel('Número de mortes')
```

Vamos agora trabalhar com o dataset Migrants.csv. O primeiro passo é entender esse dataset, então vamos exibir alguns registros inicialmente.

```python
df2 = pd.read_csv('migrants.csv', encoding='latin-1')
```

```python
df2
```

Vamos ver as diferentes regiões onde os incidentes aconteceram.

```python
df2['Region_of_Incident'].unique()
```

Queremos selecionar todas as regiões e incidente, com exceção da North America, South America, East Asia, South Asia

```python
df3 = df2[~df2['Region_of_Incident'].isin(['North America', 'South America', 'East Asia', 'South Asia'])]
df3['Region_of_Incident'].unique()
```

Agora vamos criar um gráfico de barras mostrando o número de mortos e desaparecidos entre 2014 e 2016.

```python
df3 = pd.crosstab(df3['Region_of_Incident'], df3['Reported_Year'], values=df3['Total_Dead_and_Missing'], aggfunc='sum')
df3
```

```python
plt.rcParams['figure.figsize'] = [10, 8]
sns.set_context('notebook', font_scale=1.5, rc={'font.size': 14, 'axes.titlesize': 16, 'axes.labelsize': 12})
df3.plot(kind='bar', stacked=False)
plt.title('Migrantes mortos ou desaparecidos entre 2014 e 2016')
plt.ylabel('Mortos ou desaparrecidos')
plt.xlabel('Região')
plt.show()
```
