Nesta atividade avaliativa você usará o dataset 'carros_usados', que traz informações sobre carros usados disponíveis para venda em diversos estados americanos.

```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('carros_usados.csv')
df.head()
```

### Questão 1
Crie um gráfico de barras que mostre o preço médio dos carros nacionais ('Domestic') e importados ('Import'), por categoria (vehicle.type).

```python
plt.rcParams['figure.figsize'] = [12, 8]
sns.set_style('darkgrid')
summary = pd.crosstab(df['vehicle.type'], df['domestic.import'], values=df['total.cost'], aggfunc=np.mean)
summary.plot(kind='bar', stacked=False)
plt.title('Preço médio de carros nacionais e importados por tipo de veículo')
plt.xlabel('Tipo de Veículo')
plt.ylabel('Preço Médio')
plt.show()
```

### Questão 2
Crie um gráfico de pizza que mostre a porcentagem de carros a venda do estado da California (CA) por cor (color.set).

```python
plt.rcParams['figure.figsize'] = [12, 8]
sns.set_style('white')

summary = pd.crosstab(df['color.set'], df['state'])
summary['CA'].plot(kind='pie', autopct='%1.1f%%', legend=False, subplots=True, shadow=True, startangle=240)

plt.title('Cores dos carros da Califórnia')
plt.show()
```

### Questão 3
Crie um gráfico de dispersão que mostre a relação entre a idade do veículo (vehicle.age) e seu preço (total.cost). Considere apenas veículos do tipo FAMILY.LARGE.

```python
dfQuestao3 = df[df['vehicle.type'] == "FAMILY.LARGE"]

sns.set_style('darkgrid')
sns.regplot(x='vehicle.age', y='total.cost', fit_reg=True, data=dfQuestao3)

plt.title('Relação entre idade e preços de veículo tipo FAMILY.LARGE')
plt.ylabel('Preço do veículo')
plt.xlabel('Idade do veículo')
plt.show()
```

### Questão 4
Crie um gráfico de barras que mostre o preço médio dos modelos (make.model) de carro da Toyota.

```python
plt.rcParams['figure.figsize'] = [8, 6]
dfQuestao4 = df[(df['make.model'] == "TOYOTA.AVALON") | 
                (df['make.model'] == "TOYOTA.CAMRY") | 
                (df['make.model'] == "TOYOTA.COROLLA") | 
                (df['make.model'] == "TOYOTA.TERCEL")]

sns.set_style('darkgrid')
sns.set_context('notebook', font_scale=1, rc={'font.size': 14, 'axes.titlesize': 16, 'axes.labelsize': 16})

summary = pd.crosstab(dfQuestao4['make.model'], dfQuestao4['make.model'], values=dfQuestao4['total.cost'], aggfunc=np.mean)
summary.plot(kind='bar', stacked=False)

plt.title('Preço médio dos carros da Toyota')
plt.xlabel('Modelo')
plt.ylabel('Preço')
plt.show()
```

### Questão 5
Crie um gráfico de barras que mostre o valor dos carros mais caros de cada tipo de veículo (vehicle.type) das marcas japonesas, que são: 'HONDA', 'MAZDA', 'MITSUBISHI', 'NISSAN' e 'TOYOTA'.

```python
plt.rcParams['figure.figsize'] = [8, 6]
dfQuestao5 = df[(df['makex'] == "HONDA") | 
                (df['makex'] == "MAZDA") | 
                (df['makex'] == "MITSUBISHI") | 
                (df['makex'] == "NISSAN")| 
                (df['makex'] == "TOYOTA")]

sns.set_style('darkgrid')
sns.set_context('paper', font_scale=1, rc={'font.size': 14, 'axes.titlesize': 16, 'axes.labelsize': 16})

summary = pd.crosstab(dfQuestao5['makex'], dfQuestao5['vehicle.type'], values=dfQuestao5['total.cost'], aggfunc=np.max)
summary.plot(kind='bar', stacked=False)

plt.title('Carros mais caros das marcas japonesas por tipo de veículo')
plt.xlabel('Modelo')
plt.ylabel('Preço')
plt.show()
```

### Questão 6
Crie dois histogramas lado a lado que mostrem a distribuição de preços dos veículos das marcas HONDA e HYUNDAI. Use axvline = 5000 e bins = 12.

```python
plt.rcParams['figure.figsize'] = [10, 6]
sns.set_style('darkgrid')

plt.subplot(1,2,1)
dfHonda = df[df['makex'] == "HONDA"]
sns.histplot(dfHonda['total.cost'], bins = 12, kde=False)
plt.title('Preços dos carros da Honda')
plt.xlabel('Preço')
plt.ylabel('Count')
plt.axvline(5000, color='k', linestyle='--')

plt.subplot(1,2,2)
dfHyunday = df[df['makex'] == "HYUNDAI"]
sns.histplot(dfHyunday['total.cost'], bins = 12, kde=False)
plt.title('Preços dos carros da Hyundai')
plt.xlabel('Preço')
plt.ylabel('Count')
plt.axvline(5000, color='k', linestyle='--')

plt.show()
```

### Questão 7
O dataset contém a coluna 'vehicle.age.group' que agrupa os carros de acordo com a idade. Crie dois gráficos de pizza lado a lado que exibam a porcentagem de veículos de cada grupo de idade para os estados da Califórnia (CA) e da Georgia (GA).

```python
plt.rcParams['figure.figsize'] = [12, 8]
sns.set_style('white')

summary = pd.crosstab(df['vehicle.age.group'], df['state'])
plt.subplot(1,2,1)
summary['CA'].plot(kind='pie', autopct='%1.1f%%', legend=False, subplots=True, shadow=True, startangle=240)
plt.title("Porcentagem de carros por classe - CA")
plt.subplot(1,2,2)
summary['GA'].plot(kind='pie', autopct='%1.1f%%', legend=False, subplots=True, shadow=True, startangle=240)
plt.title("Porcentagem de carros por classe - GA")
plt.tight_layout()
plt.show()
```
